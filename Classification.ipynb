```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
data = load_iris()
target = data.target
```

```python
plt.hist(target ,color='blue', orientation='vertical')
plt.xlabel('sxd1182_X_Axis')
plt.ylabel('sxd1182_Y_Axis')
plt.title('sxd1182_Histogram_Chart_Classification_For_Target_Variable')
plt.show()
```

```python
sxd1182_X_train, sxd1182_X_test, sxd1182_y_train, sxd1182_y_test = train_test_split(data.data, data.target, test_size=0.1, random_state=42)
```

```python
# Linear Discriminant Analysis
sxd1182_components = 2
sxd1182_linear_discriminant_analysis = LinearDiscriminantAnalysis(n_components=sxd1182_components)
sxd1182_linear_discriminant_analysis.fit(sxd1182_X_train, sxd1182_y_train)
sxd1182_lda_predictions = sxd1182_linear_discriminant_analysis.predict(sxd1182_X_test)
sxd1182_lda_accuracy = accuracy_score(sxd1182_lda_predictions, sxd1182_y_test)

# Naive Bayes
sxd1182_naive_bayes = GaussianNB()
sxd1182_naive_bayes.fit(sxd1182_X_train, sxd1182_y_train)
sxd1182_nb_predictions = sxd1182_naive_bayes.predict(sxd1182_X_test)
sxd1182_nb_accuracy = accuracy_score(sxd1182_nb_predictions, sxd1182_y_test)

# Logistic Regression
sxd1182_logistic_regression = LogisticRegression()
sxd1182_logistic_regression.fit(sxd1182_X_train, sxd1182_y_train)
sxd1182_logreg_predictions = sxd1182_logistic_regression.predict(sxd1182_X_test)
sxd1182_logreg_accuracy = accuracy_score(sxd1182_logreg_predictions, sxd1182_y_test)

print("Linear Discriminant Analysis Accuracy:", sxd1182_lda_accuracy)
print("Naive Bayes Accuracy:", sxd1182_nb_accuracy)
print("Logistic Regression Accuracy:", sxd1182_logreg_accuracy)
```
